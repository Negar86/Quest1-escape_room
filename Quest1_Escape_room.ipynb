```python
game_room = {'items': ['Couch', 'Piano'],'doors':['door A'] ,'keys': ['key A']}
bedroom1 = {'items': ['Qeen bed'],'doors':['door A','door B','door C'] ,'keys': ['key B']}
bedroom2 = {'items': ['Double bed','Dresser'],'doors':['door B'] ,'keys': ['key C','key D']}
living_room = {'items':['Dining table'],'doors':['door C','door D'],'keys':[]}

wallet = set()


def escape_room():
    print(f"Hello. You are in a Escape Room and have only 5 minutes to escape, otherwie something bad will happen to you. HURRY UP!!!\n\nThis room is GAME ROOM and you can see {game_room.get('items')} and {game_room.get('doors')}\n")
    
    current_room = "game_room"
    while current_room != "outside":

        if current_room == "game_room":
            ans = input("Do you want to Explore or Examine? \n1. Explore\n2. Examine\n").lower().strip()

            if ans == "1":
                ans_exp = input(f"Which item do you want to explore? {game_room.get('items')}\n").lower().strip()

                if ans_exp == "piano":
                    if "key A" in wallet:
                        print("There is nothing new here!")
                    else:
                        print("You find Key A.\n")
                        wallet.add("key A")
                elif ans_exp == "couch":
                    print("There is nothing there!\n")
                else:
                    print("please enter the item name correctly.")

            elif ans == "2":

                ans_exa = input(f"You have {wallet}.\nWhich Door you want to examine?{game_room.get('doors')}: ").lower().strip()
                if ans_exa == "door a":
                    if "key A" in wallet:
                        print(f"\nYou have entered BEDROOM1.\nYou can see {bedroom1.get('items')} and {bedroom1.get('doors')}.\n")
                        current_room = "bedroom1" 
                    else:
                        print("You need key A to unlock the door.\n")

            else:
                print("Invalid option. Please enter '1' for Explore or '2' for Examine.\n")


        if current_room == "bedroom1":
            ans = input("Do you want to explore or examine? \n1. Explore\n2. Examine\n:")

            if ans == "1":
                ans_exp = input(f"Which item do you want to explore? {bedroom1.get('items')}:\n").lower().strip()

                if ans_exp == "qeen bed":
                    if "key B" in wallet:
                        print("There is nothing new here!")
                    else:
                        print("You find key B!\n")
                        wallet.add("key B")
                else:
                    print("please enter the item name correctly.")

            elif ans == "2":

                ans_exa = input(f"You have {wallet}.\nWhich door you want to examine? {bedroom1.get('doors')}: ").lower().strip()
                if ans_exa == "door b":
                    if "key B" in wallet:
                        print(f"\nYou have entered BEDROOM 2.\nYou can see {bedroom2.get('items')} and {bedroom2.get('doors')}.\n")
                        current_room = "bedroom2" 
                    else:
                        print("You need key B to unlock the door.\n\n")

                elif ans_exa == "door a":
                    if "key A" in wallet:
                        ask= input("Are you sure you want to go back to 'GAME ROOM'???(YES/NO)\n")
                        if ask.lower().strip()=="yes":
                            print("You are back in GAME ROOM.")
                            current_room = "game_room"
                        else:
                            current_room = "bedroom1"


                elif ans_exa == "door c":
                    if "key C" in wallet:
                        print(f"\nYou have entered LIVING ROOM.\nYou can see {living_room.get('items')} and {living_room.get('doors')}.\n")
                        current_room = "living_room"             
                    else:
                        print("You need key C to unlock the door.\n")

                else:
                    "Please enter the door name correctly."

            else:
                print("Invalid option. Please enter '1' for Explore or '2' for Examine..")

        if current_room == "bedroom2":
            ans = input("Do you want to explore or examine? \n1. Explore\n2. Examine\n:").lower().strip()

            if ans == "1":
                ans_exp = input(f"Which item do you want to explore? {bedroom2.get('items')}:\n").lower().strip()

                if ans_exp == "double bed":
                    if "key C" in wallet:
                        print("There is nothing new here!")
                    else:
                        print("You find key C!\n")
                        wallet.add("key C")


                elif ans_exp == "dresser":
                    if "key D" in wallet:
                        print("There is nothing new here!")
                    else:
                        print("You find key D!\n")
                        wallet.add("key D")
                else:
                    print("please enter the item name correctly.")

            elif ans == "2":

                ans_exa = input(f"You have {wallet}.\nWhich door you want to examine? {bedroom2.get('doors')}: ")
                if ans_exa.lower().strip() == "door b":
                    if "key B" in wallet:
                        print("\nYou have entered BEDROOM 1.\n")
                        current_room = "bedroom1" 
                    else:
                        print("You need key B to unlock door_b.\n")


            else:
                print("Invalid option. Please enter '1' for Explore or '2' for Examine.")


        if current_room == "living_room":
            ans = input("Do you want to Explore or Examine? \n1. Explore\n2. Examine\n:")

            if ans == "1":
                ans_exp = input(f"Which item do you want to explore? {living_room.get('items')}:\n").lower().strip()

                if ans_exp == "dining table":
                    print("There is nothing there.")
                else:
                    print("please enter the item name correctly.")

            elif ans == "2":

                ans_exa = input(f"You have {wallet}.\nWhich door you want to examine?{living_room.get('doors')}: ").lower().strip()
                if ans_exa == "door d":
                    if "key D" in wallet:
                        print("\nHOORAY. You have escaped the room and now you are FREE.\nThank you for playing the Escape Room game!")
                        current_room = "outside" 
                    else:
                        print("You need key D to unlock the door.\n")

                elif ans_exa == "door c":
                    print("\nYou have entered BEDROOM 1 again.\n")
                    current_room = "bedroom1"

            else:
                print("Invalid option. Please enter '1' for Explore or '2' for Examine.\n")

```
